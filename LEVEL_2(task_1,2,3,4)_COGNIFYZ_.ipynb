#Task1- Analyze the distribution of aggregate ratings and determine the most common rating range and calculate the average number of votes received by restaurants.

```python
# Analyzing the distribution of aggregate ratings..
plt.figure(figsize=(10, 5))
sns.histplot(data['Aggregate rating'], bins=20, kde=True)
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Aggregate Ratings')
plt.show()
```

```python
#Determining the most common rating range..
rating_ranges = pd.cut(data['Aggregate rating'], bins=[0, 2, 3, 4, 5], labels=['0-2', '2-3', '3-4', '4-5'])
common_rating_range = rating_ranges.mode().values[0]
print("Most Common Rating Range:", common_rating_range)
```

```python
#Calculating the average number of votes received by restaurants..
average_votes = data['Votes'].mean()
print(f"The average number of votes received by restaurants is: {average_votes:.2f}")
```

# Task2- identify the most common combinations of cuisines in the dataset and determine if certain cuine combinations tend to have higher ratings.

```python
#Identifying the most common combinations of cuisines..
Cuisines = data["Cuisines"]
Aggregate_ratings = data["Aggregate rating"]
Cuisine_combinations_counts = Cuisines.value_counts()
Cuisine_combinations_counts
```

# Task3- Plot the locations of restaurants on a map using longitude and latitude coordinates and identify any patterns or cluster of restaurants in specific area.

```python
from sklearn.cluster import KMeans
```

```python
location_data = data[['Latitude', 'Longitude']]
inertia_values = []
k_range = range(1, 11)
```

```python
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(location_data)
    inertia_values.append(kmeans.inertia_)
```

```python
# Plotting the locations of restaurants on a map using longitude and latitude coordinates..
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia_values, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-cluster Sum of Squares (Inertia)')
plt.title('Elbow Curve for Optimal K')
plt.show()
```

```python
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
data['Cluster'] = kmeans.fit_predict(location_data)
```

```python
# identifying any patterns or cluster of restaurants in specific area.
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Longitude', y='Latitude', hue='Cluster', data=data, palette='Set1', s=100)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Clustering of Restaurants')
plt.legend()
plt.show()
```

# Task 4-Identify if there are any restaurant chains present in the dataset. Analyze the ratings and popularity of different restaurant chains.

```python
# Identifying if there are any restaurant chains present in the dataset..
chain_counts = data['Restaurant Name'].value_counts()
restaurant_chains = chain_counts[chain_counts > 1]
```

```python
print('Restaurant Chains in the Dataset:')
print(restaurant_chains)
```

```python
# Analyzing the ratings and popularity of different restaurant chains..
chain_info = data.groupby('Restaurant Name').agg({'Aggregate rating': 'mean', 'Votes': 'sum'})

chain_info = chain_info.sort_values(by='Aggregate rating', ascending=False)

chain_info.head(10)
```

```python
plt.figure(figsize=(12, 6))
top_chain_info = chain_info.head(10)
sns.barplot(x=top_chain_info['Aggregate rating'], y=top_chain_info.index,palette='Blues_d')
plt.xlabel('Aggregate rating')
plt.ylabel('Restaurant Chain')
plt.title('Top-Rated Restaurant Chains')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
popular_chain_info = chain_info.sort_values(by='Votes', ascending=False).head(10)
sns.barplot(x=popular_chain_info['Votes'], y=popular_chain_info.index, palette='Greens_d')
plt.xlabel('Total Votes')
plt.ylabel('Restaurant Chain')
plt.title('Most Popular Restaurant Chains')
plt.show()
```
